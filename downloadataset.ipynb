```python
import pandas as pd

# File paths
splits = {
    'train': 'credit_card_transaction_train.csv',
    'test': 'credit_card_transaction_test.csv'
}

# Read both datasets
df_train = pd.read_csv("hf://datasets/pointe77/credit-card-transaction/" + splits["train"])
df_test = pd.read_csv("hf://datasets/pointe77/credit-card-transaction/" + splits["test"])

# Combine them
df = pd.concat([df_train, df_test], ignore_index=True)
```
